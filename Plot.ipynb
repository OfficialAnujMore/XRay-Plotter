```python
data = image.imread('sunset.jpg')
  
#
# plt.plot(100, 100, marker='o', color="red")
# plt.imshow(data)
# plt.show()

#    Start, Stop
x = [50, 200]
#    x1, x2
y = [125, 100]
#    y1, y2
plt.plot(x, y, color="white", linewidth=3)
xVal = (x[1]**2-x[0]**2)
yVal = (y[1]**2-y[0]**2)

print(xVal)
print(yVal)
ans = math.sqrt(xVal+yVal)
print(ans)
plt.imshow(data)
plt.show()
```
